```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
x = np.random.uniform(low=-2, high=2, size=100)
e = np.random.normal(loc=0, scale=1, size=100)
y = 2*(x**3) + x**2 - 2*x + 5 + e
```

```python
plt.plot(x,y, '.')
plt.xlabel('x') ; plt.ylabel('y'); plt.title('y vs x')
plt.show()
```

```python
#AIC
def AIC(sse, q):
    return 2*(q+1) - 2*np.log(sse)

#BIC
def BIC(sse, N, q):
    return N*np.log(sse/N) + (q+1)*np.log(N)
```

```python
def regression(x, y, q):
    N = len(y)
    X = np.power(x.reshape(-1,1), np.arange(q+1))
    alpha = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))
    y_pred = np.dot(X, alpha)
    e = y - y_pred
    
    sse = np.sum(e**2)
    mse = (sse/N)**(0.5)
    aic = AIC(sse, q)
    bic = BIC(sse, N, q)
    
    H = np.dot(X, np.dot(np.linalg.inv(np.dot(X.T, X)), X.T))
    h = np.array([H[i][i] for i in range(N)])
    loocv = (1/N)*np.sum((e/(1-h))**2)
    
    return (mse, aic, bic, loocv)
```

```python
Q = [2, 5, 10, 20]
mse = []
aic = []
bic = []
loocv = []
for q in Q:
    tuple_ = regression(x,y,q=q)
    mse.append(tuple_[0])
    aic.append(tuple_[1])
    bic.append(tuple_[2])
    loocv.append(tuple_[3])
```

```python
def plotit(Q, y, xlab, ylab, tit):
    plt.plot(Q, y, 'o-')
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.title(tit)
    plt.show()
```

```python
plotit(Q, mse, 'q', 'MSE', 'MSE vs q')
```

```python
plotit(Q, aic, 'q', 'AIC', 'AIC vs q')
```

```python
plotit(Q, bic, 'q', 'BIC', 'BIC vs q')
```

```python
plotit(Q, loocv, 'q', 'LOOCV', 'LOOCV vs q')
```

```python
# Cross Validation
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

def crossVal(x, y, q, k_fold):
    N = len(y)
    X = np.power(x.reshape(-1,1), np.arange(q+1))
    lr = LinearRegression(fit_intercept=False)
    
    return np.mean(1 - cross_val_score(lr, X, y, cv=k_fold))
    
```

```python
#2-fold CV
K = 2
two_fcv = []
for q in Q:
    two_fcv.append(crossVal(x,y,q,K))
    
plotit(Q, two_fcv, 'q', '2FCV', '2FCV vs q')
```

```python
#3-fold CV
K = 3
three_fcv = []
for q in Q:
    three_fcv.append(crossVal(x,y,q,K))
    
plotit(Q, three_fcv, 'q', '3FCV', '3FCV vs q')
```
